Чёрный куб покрасили снаружи белой краской, затем разрезали на 27 одинаковых маленьких кубиков и как попало сложили из них большой куб. С какой вероятностью все грани этого куба будут белыми?

Каждый кубик может быть как поставлен в любое место в большом кубе (**позиция**) - 27 вариантов, так и иметь определённое положение относительно грани куба (**вращение**) - 24 варианта (данные параметры не взаимосвязаны).

Общее количество вариантов размещения кубиков = $ (27*24) * (26*24) * ... * (1*24) = 27! * 24^{27} $

Определим группы кубиков после окраски и разрезки (данные типы не взаимосвязаны, т.е. окончательное количество вариантов можно определить, как произведение вариантов для каждого типа):
1. полностью чёрный кубик из центра большого.
2. кубики с одной белой гранью, количество = 6 (количество граней куба, кубик взят из центра грани).
3. кубики с двумя белыми гранями, количество = 12 (располагаются в центрах рёбер куба).
4. кубики с тремя белыми гранями, количество = 8 (располагаются по углам куба).

Определим количество вариантов размещения кубиков различного типа для получения искомого куба:
1. 24 варианта вращений и 1 позиция = $ 1! * 24^1 = 24 $ варианта
2. для кубиков с одной белой гранью есть 4 варианта вращений и 6 позиций, т.е. $ 6! * 4^6 $ вариантов.
3. с двумя белыми гранями - 2 варианта вращений и 12 позиций: $ 12! * 2^{12} $ вариантов
4. с тремя - 3 варианта вращений и 8 позиций: $ 8! * 3^8 $ вариантов.

Тогда искомое количество подходящих комбинаций равно $ 24 * 6! * 4^6 * 12! * 2^{12} * 8! * 3^8  = 24 * 6! * 8! * 12! * 2^{24} * 3^8 $

А вероятность сборки куба со всем белыми гранями равна $ \frac{24 * 6! * 8! * 12! * 2^{24} * 3^8}{27! * 24^{27}} = \frac{6! * 8! * 12! * 2^{24} * 3^8}{27! * 24^{26}} = \frac{6! * 8! * 12!}{27! * 6^6 * 8^8 * 12^{12}} $

Искомая величина имеет порядок $ 10^{-37} $, что практически равно 0.

Определим вероятность построения куба со всеми белыми гранями через вероятность установки кубиков каждого типа в подходящую позицию.
1. вероятность выбора чёрного кубика и установки его в подходящую позицию будет равна $ \frac{1}{27} $
2. из оставшихся 26 кубиков 6 раз подряд выбираем кубики с одной белой гранью и устанавливаем в нужные позиции (положение одной грани из 6). вероятность этого события = $ \prod_{i=0}^5 \frac{6 - i}{26 - i} * \frac{1}{6} = \frac{1}{6^6} * 6! * \prod_{i=0}^5 \frac{1}{26 - i} $
3. из оставшихся 20 кубиков 12 раз подряд выбираем кубики с двумя белыми гранями и устанавливаем в нужные позиции (положение двух граней из 24 вариантов вращения). искомая вероятность = $ \prod_{i=0}^{11} \frac{12 - i}{20 - i} * \frac{1}{12} = \frac{1}{12^{12}} * 12! * \prod_{i=0}^{11} \frac{1}{20 - i}$
4. оставшиеся 8 кубиков с тремя гранями располагаем в нужном положении (положение 3 граней из 24 вариантов). вероятность этого события равна $ \prod_{i=1}^8\frac{1}{8} = \frac{1}{8^8}$

Далее, перемножаем эти вероятности: раскрывая произведения, их можно свести к виду $ \frac{8!}{27!} $ (в общем знаменателе произведений присутствует произведение от 9 до 27 включительно), тогда общая вероятность равна $ \frac{6! * 8! * 12!}{27! * 6^6 * 8^8 * 12^{12}} $
