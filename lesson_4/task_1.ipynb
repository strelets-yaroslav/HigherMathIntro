1. Решите уравнение 

sin(x)/x=0

$ \frac{sin(x)}{x} = 0 $

$ sin(x) = 0 $

$ x = \pi * n, n \in \mathbb{Z}, n \neq 0 $

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
start, end, n = -np.pi*10, np.pi*10, 501
x = np.linspace(start, end, n)
plt.plot(x, np.sin(x) / x)
plt.plot([start, end], [0, 0], color='r', linestyle='--', linewidth=0.5)
for i in np.arange(start, end, np.pi): # цикл по pi*n
    if i != 0: # исключаем ноль, т.к. в ней получаем бесконечность
        plt.plot([i, i], [0, 1], color='b', linestyle=':', linewidth=0.5)
plt.show()
```
