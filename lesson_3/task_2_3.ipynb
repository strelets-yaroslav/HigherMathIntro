```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def polar2decart(r, alpha):
    return r * np.cos(alpha), r * np.sin(alpha)


polar2decart(10, np.pi / 3)
```

```python
alpha = np.arange(0., 2., 1./180.)*np.pi
r = 5

plt.polar(alpha, [r]*len(alpha)) 
plt.show()
```

```python
a = [np.pi/6, np.pi/2]
b = [2, 6]

plt.polar(a, b)
plt.show()
```
