2. Задание 

Докажите, что при ортогональном преобразовании сохраняется расстояние между точками.

Определим расстояние между точками, как длину вектора вида **|xy|** с координатами $ (x_1, y_1), (x_2, y_2) $

Преобразование в общем виде: $ X = a_{11}*x + a_{12} * y + a_{13}, Y = a_{21} * x + a_{22} * y + a_{23} $

$ (X_2 - X_1) = a_{11}*x_2 + a_{12}*y_2 + a_{13} - a_{11}*x_1 + a_{12}*y_1 + a_{13} = a_{11}*(x_2 - x_1) + a_{12}*(y_2 - y_1) $, для Y аналогично

Тогда длина преобразованного вектора **|XY|** (для простоты берём квадрат длины):

$ |XY|^2 = (X_2 - X_1)^2 + (Y_2 - Y_1)^2 = (a_{11}*(x_2 - x_1) + a_{12}*(y_2 - y_1))^2 + (a_{21}*(x_2 - x_1) + a_{22}*(y_2 - y_1))^2 $

После раскрытия скобок и объединения множителей получаем:

$ |XY|^2 = (a_{11}^2 + a_{21}^2)*(x_2 - x_1)^2 + (a_{12}^2 + a_{22}^2)*(y_2 - y_1)^2 +2*(a_{11}*a_{12} + a_{21}*a_{22})*(x_2 - x_1)*(y_2 - y_1) $

Учитывая свойства ортогонального преобразования: $ a_{11}^2 + a_{21}^2 = 1, a_{12}^2 + a_{22}^2 = 1, a_{11}*a_{12} + a_{21}*a_{22} = 0 $, получаем:

$ |XY|^2 = (x_2 - x_1)^2 + (y_2 - y_1)^2 = |xy|^2 $, т.е. **|XY| = |xy|**
