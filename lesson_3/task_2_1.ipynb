Нарисуйте график функции: $ k * cos(x - a) + b $ для некоторых (2-3 различных) значений параметров k, a, b

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def param_cos(x, k, a, b):
    return k * np.cos(x - a) + b


x = np.linspace(-3*np.pi, 5*np.pi, 501)

k1, a1, b1 = 6, 11, 20
k2, a2, b2 = 3, 15, 25
k3, a3, b3 = 2, 4, 7

plt.figure(figsize = (10, 5))
plt.plot(x, param_cos(x, k1, a1, b1), color='g', label='k1, a1, b1')
plt.plot(x, param_cos(x, k2, a2, b2), color='r', label='k2, a2, b2')
plt.plot(x, param_cos(x, k3, a3, b3), label='k3, a3, b3')

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend(frameon=False)
plt.show()
```
