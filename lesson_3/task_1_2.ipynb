Почему прямые не кажутся перпендикулярными? (см.ролик)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 21)
y1 = 3 * x + 1
y2 = (-1/3) * x + 1
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel("x")
plt.ylabel("y")
```

Из-за того, что координатная сетка не отмасштабирована, а также окно графика растянуто, создаётся впечатление, что прямые не перпендикулярны.

```python
a, b = -20, 20

plt.figure(figsize = (10, 10)) # выравниваем окно для графика
x = np.linspace(a, b)
y1 = 3 * x + 1
y2 = (-1/3) * x + 1
plt.plot(x,y1)
plt.plot(x,y2)
plt.ylim(a, b) # синхронизируем масштабирование
plt.xlim(a, b)
```
