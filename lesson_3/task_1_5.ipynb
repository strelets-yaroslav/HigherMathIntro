Нарисуйте трехмерный график двух параллельных плоскостей.

Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
import numpy as np
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

трехмерный график двух параллельных плоскостей.

```python
fig = figure(figsize = (8, 8))
ax = Axes3D(fig)

start, end = -20, 20

X = np.arange(start, end, 2)
Y = np.arange(start, end, 2)
X, Y = np.meshgrid(X, Y)

A, B, C, D1, D2 = -5, 6.2, 8, 100, -15

Z1 = (-A * X - B * Y - D1) / C
Z2 = (-A * X - B * Y - D2) / C

ax.plot_wireframe(X, Y, Z1, label='Ax + By + Cz + D1 = 0', linestyle='-.', linewidth=1)
ax.plot_wireframe(X, Y, Z2, colors='orange',label='Ax + By + Cz + D2 = 0')
plt.legend(frameon=False)
show()
```

трехмерный график двух любых поверхностей второго порядка.

```python
fig = figure(figsize = (8, 8))
ax = Axes3D(fig)

a1, b1, a2, b2 = 1, 2, 2, 1

start, end = -20, 20

X = np.arange(start, end, 2)
Y = np.arange(start, end, 2)
X, Y = np.meshgrid(X, Y)

f = lambda x, y: x ** 2 / a1 ** 2 - y ** 2 / b1 ** 2

Z1 = f(X, Y)
Z2 = X**2 / a2**2 + Y**2 / b2**2 

ax.plot_surface(X, Y, Z1,  linewidth=1, color='r')
ax.plot_surface(X, Y, Z2,  linewidth=1, color='b')
show()
```
