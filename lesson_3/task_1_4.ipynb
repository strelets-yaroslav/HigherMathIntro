2)  Пусть задана плоскость: A1x + B1y + C1z + D1 = 0 и прямая: $ \frac{x - x_1}{x_2 - x_1} = \frac{y - y_1}{y_2 - y_1} = \frac{z - z_1}{z_2 - z_1} $

Как узнать, принадлежит прямая плоскости или нет?

1. Для прохождения плоскости через начало координат, нужно, чтобы для точки с координатами (0, 0, 0) выполнялось равенство Ax + By + Cz + D = 0, следовательно, D должен быть равен 0, и тогда уравнение плоскости будет Ax + By + Cz = 0

```python
import numpy as np
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure(figsize = (8, 8))
ax = Axes3D(fig)

start, end = -20, 20

X = np.arange(start, end, 2)
Y = np.arange(start, end, 2)
X, Y = np.meshgrid(X, Y)

A, B, C, D = 3, 0.4, -2, 100

Z1 = (-A * X - B * Y - D) / C
Z2 = (-A * X - B * Y) / C

ax.plot_wireframe(X, Y, Z1, label='Ax + By + Cz + D = 0', linestyle='-.', linewidth=1)
ax.plot_wireframe(X, Y, Z2, colors='orange',label='Ax + By + Cz = 0')
ax.scatter(0, 0, 0, 'z', 50, 'red', label='Начало координат (0, 0, 0)')
plt.legend(frameon=False)
show()
```

2. В уравнении прямой мы имеем координаты двух точек, через которые проходит прямая, (x1, y1, z1) и (x2, y2, z2).
Если при подстановке этих координат в уравнение плоскости будет выполняться равенство, значит прямая принадлежит плоскости.
