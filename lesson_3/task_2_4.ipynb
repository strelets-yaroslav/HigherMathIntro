exp(x) + x∙(1 – y) = 1  

y = x2 – 1


Решите систему уравнений и неравенств:

y = x2 – 1

exp(x) + x∙(1 – y)  - 1 > 0


1. система уравнений

$ y = \frac{e^x + x - 1}{x} $

$ y = x^2 - 1 $

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
x = np.linspace(-3, 4, 201)

plt.figure(figsize = (10, 15))
plt.plot(x, (np.exp(x) + x - 1) / x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)
plt.axis('scaled')
plt.show()
```

```python
def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 = fsolve(equations, (2, 4))
x2, y2 = fsolve(equations, (-1, 1))
              
print(x1, y1)
print(x2, y2)
```

2. система уравнений и неравенств

при $ x < 0, y > \frac{e^x + x - 1}{x} $, при $ x > 0, y < \frac{e^x + x - 1}{x} $

$ y = x^2 - 1 $

$ y = -\frac{1}{x} + 3 $

```python
x = np.linspace(-3, 3, 301)

plt.figure(figsize = (10, 15))
plt.plot(x, (-1 / x) + 3)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2, 7) 
plt.grid(True)

plt.show()
```

```python
def equations(p):
    x, y = p
    return (x**2 - 1 - y, (-1 / x) + 3 - y)

x1, y1 =  fsolve(equations, (2, 2))
x2, y2 =  fsolve(equations, (0.5, -0.5))
x3, y3 =  fsolve(equations, (-1.5, 2))
              
print (x1, y1)
print (x2, y2)
print (x3, y3)
```
