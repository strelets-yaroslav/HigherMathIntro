5. Найдите нормальное псевдорешение недоопределенной системы:

x + 2y – z = 1  
8x – 5y + 2z = 12

Для этого определите функцию Q(x,y,z), равную норме решения, и найдите ее минимум. 

Выражая через х из уравнений сначала y (умножаем первое на 2 и складываем со вторым), затем z (подставляем в первое полученное выражение для у), получаем:

y = 10*x - 14  
z = 21*x - 29

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
def Q(x, y, z):
  return x**2 + y**2 + z**2


fig = figure()
ax = Axes3D(fig)
X = np.arange(0, 3, 0.01)
Y = np.arange(-10, 10, 0.1)
X, Y = np.meshgrid(X, Y)
ax.plot_surface(X, Y, Q(X, 10*X - 14, 21*X - 29))
show()
```
