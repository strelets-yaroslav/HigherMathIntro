Решите **нелинейную** систему уравнений:
$$\left\{
    \begin{array}\\
        x^2 + xy\ – 9 = 0, {нелинейное\ уравнение}\\
        x\ – \frac{y}{5} = 0, {линейное\ уравнение}\\
    \end{array}
\right.$$

Выражаем $y$ во втором уравнении через $х$: $y = 5x$ и подставляем в первое: $x^2 + 5 * x * x = 9, => 6x^2 = 9$

Отсюда $x^2 = \frac{3}{2}, => x = \pm \sqrt{\frac{3}{2}}, y = \pm 5\sqrt{\frac{3}{2}}$

```python
%matplotlib inline
from numpy import linspace, sqrt
import matplotlib.pyplot as plt


def y(x):
    return (9 - x**2) / x

x_1 = linspace(-2, -1, 100)
x_2 = linspace(1, 2, 100)

plt.figure(1)
plt.plot(x_1, y(x_1))
plt.plot(x_1, 5*x_1)
plt.axvline(-sqrt(1.5), ls='-', color="red")
plt.axhline(-5*sqrt(1.5), ls='-', color="red")

plt.figure(2)
plt.plot(x_2, y(x_2))
plt.plot(x_2, 5*x_2)
plt.axvline(sqrt(1.5), ls='-', color="red")
plt.axhline(5*sqrt(1.5), ls='-', color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.show
```
