Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
%matplotlib inline
from numpy import cos, pi, linspace
import matplotlib.pyplot as plt


def k_cos(k, x):
    return cos(k * x)


x = linspace(-3*pi, 3*pi, 300)
y_1 = k_cos(1, x)
k_2 = 3.1
y_2 = k_cos(k_2, x)

plt.plot(x, y_1, label="cos(x)")
plt.plot(x, y_2, label=f"cos({k_2}*x)")
plt.xlabel("x")
plt.ylabel("y")
legend = plt.legend(loc='upper right', shadow=True, fontsize='x-small')
legend.get_frame().set_facecolor('lightgray')
plt.show
```
